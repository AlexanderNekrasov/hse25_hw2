````python
%%shell
samtools sort -o bowtie2_res/res_ZQM_filtered_sorted.bam bowtie2_res/res_ZQM_filtered.bam
samtools sort -o bowtie2_res/res_AVO_filtered_sorted.bam bowtie2_res/res_AVO_filtered.bam
samtools sort -o bowtie2_res/res_AXD_filtered_sorted.bam bowtie2_res/res_AXD_filtered.bam
````

````python
%%shell
java -jar picard.jar MarkDuplicates \
      I=bowtie2_res/res_AVO_filtered_sorted.bam \
      O=bowtie2_res/res_AVO_filtered_dedup.bam \
      M=bowtie2_res/res_AVO_duplicate_metrics.txt \
      REMOVE_DUPLICATES=true \
      VALIDATION_STRINGENCY=LENIENT
````

````python
%%shell
java -jar picard.jar MarkDuplicates \
      I=bowtie2_res/res_AXD_filtered_sorted.bam \
      O=bowtie2_res/res_AXD_filtered_dedup.bam \
      M=bowtie2_res/res_AXD_duplicate_metrics.txt \
      REMOVE_DUPLICATES=true \
      VALIDATION_STRINGENCY=LENIENT
````

## –ü–æ–¥—Å—á—ë—Ç ngsplot –ø—Ä–∏ –ø–æ–º–æ—â–∏ deeptools
–ù–µ–æ–±—Ö–æ–¥–∏–º–æ —Å–∫–∞—á–∞—Ç—å –∞–Ω–Ω–æ—Ç–∞—Ü–∏—é (–Ω–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è hg19 –≤–∑—è—Ç—å –æ—Ç—Å—é–¥–∞ https://hgdownload-test.gi.ucsc.edu/goldenPath/hg19/bigZips/genes/ - hg19.knownGene)

–î–ª—è hg38 - https://hgdownload-test.gi.ucsc.edu/goldenPath/hg38/bigZips/genes/

````python
!wget https://hgdownload-test.gi.ucsc.edu/goldenPath/hg19/bigZips/genes/hg19.knownGene.gtf.gz
````

````python
# –ø–æ–¥—Å—á—ë—Ç –º–∞—Ç—Ä–∏—Ü—ã. –í –ø—Ä–∏–º–µ—Ä–µ —Å—á–∏—Ç–∞–µ—Ç—Å—è 30 –º–∏–Ω—É—Ç.
%%time
!computeMatrix scale-regions \
 -S ENCFF366SXK.bigWig \
 -R hg19.knownGene.gtf.gz \
 -out matrix.tab.gz \
 -a 2000 -b 2000 \
 --regionBodyLength 4000 \
 --skipZeros \
 --missingDataAsZero
````

````python
!plotHeatmap \
 -m matrix.tab.gz \
 -out result.png \
 --colorMap YlGnBu \
 --regionsLabel 'genes' \
 --heatmapHeight 15 \
 --plotTitle 'ngs_plot' --verbose
````
